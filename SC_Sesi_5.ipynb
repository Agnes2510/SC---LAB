Soal : cari turunan pertama dari Cos (x)

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-poster')
#untuk mengatur gaya dari grafiknya

%matplotlib inline
#untuk menampilkan grafik

h = 0.2
#h adalah jarak atau selisih antar x
x = np.arange(0, 2*np.pi, h)
#Membuat array x dari 0 sampai 2 dengan jarak antar elemen sebesar h. Ini adalah domain dari fungsi.
y = np.cos(x)
#hitung nilai cos(x) dalam tiap titik dalam array

forward_diff = np.diff(y)/h
#hitung turunan dengan metode forward difference
x_diff = x[:-1]
#Karena hasil forward_diff lebih pendek satu elemen daripada x, maka kita buang titik terakhir dari x agar cocok ukurannya untuk plotting.

exact_solution = -np.sin(x_diff)
#Menghitung turunan eksak dari cos(), yaitu sin(x), pada titik-titik yang sesuai (x_diff).

#untuk membuat grafik nya menggunakan garis "--" untuk numerik
plt.figure(figsize = [12, 8])
plt.plot(x_diff, forward_diff,'--', label = 'Finite difference approx')
plt.plot(x_diff, exact_solution, label = 'Exact Solution')
plt.legend()
plt.show()

#Menghitung galat maksimum (maximum error) antara hasil pendekatan
#numerik dan solusi eksak. Galat dihitung sebagai selisih absolut maksimum dari kedua array tersebut.
max_error = max(abs(exact_solution - forward_diff))
print(max_error)
```

```python
#riemann

import numpy as np
def f(x):
  return 4*x**4 + 2*x**3 - 6*x

#Left riemann
a = -3
b = 3
n = 30

lebar = (b-a)/(n-1)
x = np.linspace(a, b, n)
y = f(x)

left_riemann = lebar*sum(y[:n-1])
print(left_riemann)

#Right riemann
a = -3
b = 3
n = 30

lebar = (b-a)/(n-1)
x = np.linspace(a, b, n)
y = f(x)

right_riemann = lebar*sum(y[1:])
print(right_riemann)
```

```python
#simpsons rule
a = 0
b = np.pi
n = 10
h = (b-a)/n
x = np.linspace(a, b, n+1)
y = np.sin(x)

S = h/3 * (y[0] + y[-1] + 4*sum(y[1:-1:2]) + 2*sum(y[2:-2:2]))
exact = 2
error = abs(S-exact)
print("Simpsons: ", S)
print("exact: ", exact)
print("error: ", error)
```

Modul Sesi 5 - Soal 2

```python
def f(x):
    return np.sin(x)

a = 0
b = np.pi

# Trapezoid 8 panel
n1 = 8
h1 = (b - a) / n1
x1 = np.linspace(a, b, n1 + 1)
y1 = f(x1)
trapz_8 = h1 * (y1[0] + 2 * sum(y1[1:-1]) + y1[-1]) / 2
print(f"Trapezoidal (8 panel): {trapz_8}")

# Trapezoid 16 panel
n2 = 16
h2 = (b - a) / n2
x2 = np.linspace(a, b, n2 + 1)
y2 = f(x2)
trapz_16 = h2 * (y2[0] + 2 * sum(y2[1:-1]) + y2[-1]) / 2
print(f"Trapezoidal (16 panel): {trapz_16}")
```

Error Graphic

```python
def f(x):
    return np.sin(x)

a = 0
b = np.pi
exact = 2
errors = []
ns = [2, 4, 8, 16, 32, 64, 128]

for n in ns:
    h = (b - a) / n
    x = np.linspace(a, b, n + 1)
    y = f(x)
    trapz = h * (y[0] + 2 * sum(y[1:-1]) + y[-1]) / 2
    error = abs(exact - trapz)
    errors.append(error)
    print(f"n = {n}, Integral = {trapz}, Error = {error}")

# Plot grafik error
plt.figure(figsize=(10,6))
plt.plot(ns, errors, marker='o')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Jumlah Panel (n)')
plt.ylabel('Error')
plt.title('Error Metode Trapezoid terhadap Jumlah Panel')
plt.grid(True, which='both')
plt.show()
```
