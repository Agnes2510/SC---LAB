```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-poster')

# Fungsi diferensial
def f(t, s):
    return np.exp(-t)

# Parameter untuk metode Euler
h = 0.1
t_start = 0
t_end = 1

t_values = np.arange(t_start, t_end + h, h)

# Inisialisi solusi
s_values = np.zeros(len(t_values))
s_values[0] = -1  # inisiasi nilai awal

# Metode Euler
for i in range(len(t_values) - 1):
    s_values[i + 1] = s_values[i] + h * f(t_values[i], s_values[i])

# Hasil dari plot
plt.figure(figsize=(12, 8))
plt.plot(t_values, s_values, 'bo--', label='approx (Euler h=0.1)')
plt.plot(t_values, -np.exp(-t_values), 'g', label='exact solution')
plt.legend()
plt.xlabel('t')
plt.ylabel('s(t)')
plt.title('Euler Method vs Exact Solution')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# mendeefinisikan fungsi turunan (dy/dx)
def f(x, y):
    return x**2 - 4*y

# Parameter metode Euler
h = 0.1
x_start = 0
x_end = 1

# buat array nilai x
x_values = np.arange(x_start, x_end + h, h)

# Inisialisasi array solusi y
y_values = np.zeros(len(x_values))
y_values[0] = 1  # inisialisasi nilai awal y(0) = 1

# metode Euler
for i in range(len(x_values) - 1):
    y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i])

# eksak
def exact_solution(x):
    return (x**2 / 4 - x / 8 + 1 / 32) + (31 / 32) * np.exp(-4 * x)

# HASIL
plt.figure(figsize=(12, 8))
plt.plot(x_values, y_values, 'bo--', label='Approx (Euler h=0.1)')
plt.plot(x_values, exact_solution(x_values), 'g', label='Exact Solution')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Solusi dari $y' + 4y = x^2$ dengan Metode Euler")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Mendefinisikan fungsi turunan
def f(t, y):
    return np.sin(y)

# Parameter
h = 0.1
t_start = 0
t_end = 5
y0 = 1

# Buat array t untuk Euler
t_values = np.arange(t_start, t_end + h, h)
y_values = np.zeros(len(t_values))
y_values[0] = y0

# Metode Euler
for i in range(len(t_values) - 1):
    y_values[i + 1] = y_values[i] + h * f(t_values[i], y_values[i])

# Solusi numerik pakai solve_ivp
sol = solve_ivp(f, [t_start, t_end], [y0], t_eval=t_values)

# HASIL
plt.figure(figsize=(12, 8))
plt.plot(t_values, y_values, 'bo--', label='Euler Approx (h = 0.1)')
plt.plot(t_values, sol.y[0], 'g', label='solve_ivp (Numerik Akurat)')
plt.xlabel('t')
plt.ylabel('y')
plt.title("Solusi dari $y' = \sin y$ dengan Euler vs solve_ivp")
plt.legend()
plt.grid(True)
plt.show()
```
