NEWTON INTERPOLATION

```python
%matplotlib inline
def divided_diff (x,y):

  #function to mcalclate the divided differences table
  n = len (y)
  coef = np.zeros([n, n])
  #the ifirst column is y
  coef [:, 0] = y

  for j in range (1, n):
    for i in range (n-j):
      coef [i][j] = \
      (coef[i+1][j-1] - coef [i][j-1]) / (x[i+j]-x[i])
  return coef

def newton_poly(coef, x_data, x): #untuk interpolasi newton
  n = len(x_data) - 1
  p = coef[n]
  for k in range (1, n + 1):
    p = coef[n - k] + (x - x_data[n - k]) * p
  return p
```

```python
x = np.array ([-5, -1, 0, 2])
y = np.array ([-2, 6, 1, 3])

a_s = divided_diff(x,y)[0,:]

x_new = np.arange (-5, 2.1, .1)
y_new = newton_poly (a_s, x, x_new)

plt.figure (figsize = (12, 8))
plt.plot (x, y, 'bo')
plt.plot (x_new, x_new, 'go')
plt.plot (x_new, y_new)
```

**MODUL ** (jawab dengan coding phyton linear interpolation sertakan plot grafik)
Olimpiade 2006, juara lomba sprint 100m adalah Ben. Catatan larinya ketika 0 s kecepatannya 10 m/s. Pada saat 5 meter kecepatannya 15m/s. Pada saat 10 meter kecepatannya adalah 20 m/s. Berapa kecepatan larinya saat pada jarak 7.2 meter?

```python
import numpy as np
import matplotlib.pyplot as plt

distances = np.array([0, 5, 10])
speeds = np.array([10, 15, 20])

target_distance = 7.2
estimated_speed = np.interp(target_distance, distances, speeds)

print(f"Kecepatan lari pada jarak {target_distance} meter adalah {estimated_speed:.2f} m/detik")

x_new = np.linspace(0, 10, 100)
y_new = np.interp(x_new, distances, speeds)


plt.plot(x_new, y_new, label="Interpolasi", color="blue")
plt.scatter(distances, speeds, color="red", label="Data Asli")

plt.xlabel("Jarak (meter)")
plt.ylabel("Kecepatan (m/detik)")
plt.title("Interpolasi Kecepatan Lari Berdasarkan Jarak")
plt.show()
```
